## Cleaning and sorting

```python
import pandas as pd
df = pd.DataFrame(data)
df.columns = df.iloc[0]
df = df.iloc[1:]
df = df[df['Country Name'] != "World"]
df = df.drop(['Series Code', 'Country Code'], axis=1)
df = df[df['Series Name'] != ""]
for year in range(1960, 2025):
    df= df.rename(columns={f'{year} [YR{year}]': f'{year}'})
#df.head()
df.tail()
```

### Grouping Data

```python
def fix_Series(name):
  match name:
    case "Energy use (kg of oil equivalent per capita)":
      return "Energy use per capita"
    case "Energy use (kg of oil equivalent) per $1,000 GDP (constant 2021 PPP)":
      return "Energy use per $1K GDP"
    case "GDP (constant 2015 US$)":
      return "GDP"
    case "Population, total":
      return "Population"
```

```python
grouped_dfs = df.groupby('Series Name')
series_dfs = {}  # Dictionary to store dataframes

for name, group in grouped_dfs:
  series_name = fix_Series(name)
  series_df = group.drop(['Series Name'], axis=1).reset_index(drop=True)
  series_dfs[series_name] = series_df
```

### Individual Pie charts

```python
import matplotlib.pyplot as plt
import seaborn as sns
import os
from google.colab import drive
from matplotlib.animation import FuncAnimation, PillowWriter


drive.mount('/content/drive', force_remount=True) # Add force_remount=True to explicitly request authentication
folder_name = "/content/drive/My Drive/SpinoffR&E Collab/Pies" # Update the file path to your Google Drive

for series_name in series_dfs:
  series_df = series_dfs[series_name]

  #ensure there are folders for each series and get the path for that folder
  inner_folder_name = os.path.join(folder_name, series_name.replace(" ", "_"))
  os.makedirs(inner_folder_name, exist_ok=True)

  # Create a pie chart using fig, ax
  fig, ax = plt.subplots(figsize=(10, 6))

  for year in range(1960, 2025):
    ax.clear()
    year_str = str(year)

    # remove the empty countries
    year_data = series_df[series_df[year_str].notna()]
    #print(year_data.head())

    # skip years if there is no data for that year at all
    if year_data.empty or year_data[year_str].sum()==0:
      continue

    chart_title = f'{series_name}: {year}'
    labels = year_data['Country Name']
    sizes = year_data[year_str]

    # Create pie chart on the ax object
    ax.pie(sizes, labels=labels, autopct='%1.1f%%', startangle=140)
    ax.set_title(chart_title)
    ax.axis('equal')  # Equal aspect ratio to ensure circle

    # Sanitize filename for individual years
    safe_title = "".join(c for c in chart_title if c.isalnum() or c.isspace()).rstrip()
    file_path = os.path.join(inner_folder_name, f'{safe_title}.png')

    # Save and close
    fig.savefig(file_path, bbox_inches='tight')
    plt.close(fig)

  print("Completed making pie charts for", series_name)
```

```python
import matplotlib.pyplot as plt

#animate function
def animate(year):
    ax.clear()
    year_str = str(year)

    if year_str not in series_df.columns:
      return

    # remove the empty countries
    year_data = series_df[series_df[year_str].notna()]

    # skip years if there is no data for that year at all
    if year_data.empty or year_data[year_str].sum()==0:
      return

    chart_title = f'{series_name}: {year}'
    labels = year_data.index
    sizes = year_data[year_str]

    # Create pie chart on the ax object
    ax.pie(sizes, labels=labels, autopct='%1.1f%%', startangle=140)
    ax.set_title(chart_title)
    ax.axis('equal')  # Equal aspect ratio to ensure circle
```

```python

import matplotlib.pyplot as plt
import os
from google.colab import drive
from matplotlib.animation import FuncAnimation, PillowWriter


drive.mount('/content/drive', force_remount=True) # Add force_remount=True to explicitly request authentication
folder_name = "/content/drive/My Drive/SpinoffR&E Collab" # Update the file path to your Google Drive
#ensure there is a folder for the files
os.makedirs(folder_name, exist_ok=True)

for series_name in series_dfs:
  series_df = series_dfs[series_name]

  # Create a pie chart using fig, ax
  fig, ax = plt.subplots(figsize=(10, 6))

  valid_years = []
  for year in range(1960, 2025):
    year_str = str(year)
    if year_str in series_df.columns:
      year_data = series_df[series_df[year_str].notna()]
      if not year_data.empty and year_data[year_str].sum() > 0:
        valid_years.append(year)

  if not valid_years:
    print(f"No valid years found for {series_name}. Skipping...")
    plt.close(fig)
    continue

  #animate system
  animation = FuncAnimation(fig, animate, frames=valid_years, interval=500)
  gif_path = os.path.join(folder_name, f'{series_name}.gif')
  animation.save(gif_path, writer='pillow')
  plt.close(fig)
  print("Completed making animated pie charts for", series_name)
```
