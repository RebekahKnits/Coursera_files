### Grouping Data

```python
def fix_Series(name):
  match name:
    case "Energy use (kg of oil equivalent per capita)":
      return "Energy use per capita"
    case "Energy use (kg of oil equivalent) per $1,000 GDP (constant 2021 PPP)":
      return "Energy use per $1K GDP"
    case "GDP (constant 2015 US$)":
      return "GDP"
    case "Population, total":
      return "Population"
    case "Energy use (Kg of oil)":
      return "Energy use"
```

```python
grouped_dfs = df.groupby('Series Name')
series_dfs = {}  # Dictionary to store dataframes

for name, group in grouped_dfs:
  #correct the long series names to short ones.
  series_name = fix_Series(name)

  series_df = group.set_index("Country Name").drop(['Series Name'], axis=1)

  # Convert year columns to numeric, coercing errors
  for col in series_df.columns:
      series_df[col] = pd.to_numeric(series_df[col], errors='coerce')
      series_df[col] = series_df[col].fillna(0)


  series_dfs[series_name] = series_df

  # Get the 'World' row
  world_row = series_df.loc["World"]
  series_df = series_df.drop("World")

  # Get only the country rows (i.e., exclude 'World')
  #country_rows = series_df.drop("World")

  # Sum the countries
  countries_sum = series_df.sum()

  # Calculate Rest of World (ROW)
  row_values = world_row - countries_sum

  # Add ROW as a new row to the DataFrame
  series_df.loc["ROW"] = row_values

  # Save the modified DataFrame back to the dictionary
  series_dfs[series_name] = series_df
```

```python
import matplotlib.pyplot as plt

#animate function
def animate(year):
    ax.clear()
    year_str = str(year)

    if year_str not in series_df.columns:
      return

    # remove the empty countries
    year_data = series_df[series_df[year_str].notna()]

    # skip years if there is no data for that year at all
    if year_data.empty or year_data[year_str].sum()==0:
      return

    chart_title = f'{series_name}: {year}'
    labels = year_data['Country Name']
    sizes = year_data[year_str]

    pie_colors = [country_colors[country] for country in labels]

    # Create pie chart on the ax object
    #ax.pie(sizes, labels=labels, colors = pie_colors, autopct='%1.1f%%', startangle=140)
    ax.pie(sizes, colors = pie_colors, startangle=140)
    ax.set_title(chart_title)
    ax.axis('equal')  # Equal aspect ratio to ensure circle

    # Add the legend to your plot
    ax.legend(handles=legend_elements, loc='upper right', bbox_to_anchor=(1.05, 1.0))
```

```python
import matplotlib.pyplot as plt
import os
from google.colab import drive
from matplotlib.animation import FuncAnimation, PillowWriter


drive.mount('/content/drive', force_remount=True) # Add force_remount=True to explicitly request authentication
folder_name = "/content/drive/My Drive/SpinoffR&E Collab" # Update the file path to your Google Drive
#ensure there is a folder for the files
os.makedirs(folder_name, exist_ok=True)

for series_name in series_dfs:
  series_df = series_dfs[series_name]

  # Create a pie chart using fig, ax
  fig, ax = plt.subplots(figsize=(12, 6))

  valid_years = []
  for year in range(1990, 2023):  # Pop and GDP go to 1960-2022
    year_str = str(year)
    if year_str in series_df.columns:
      year_data = series_df[series_df[year_str].notna()]
      if not year_data.empty and year_data[year_str].sum() > 0:
        valid_years.append(year)

  if not valid_years:
    print(f"No valid years found for {series_name}. Skipping...")
    plt.close(fig)
    continue

  #animate system
  animation = FuncAnimation(fig, animate, frames=valid_years, interval=500)
  gif_path = os.path.join(folder_name, f'g_{series_name}.gif')
  animation.save(gif_path, writer='pillow')
  plt.close(fig)
  print("Completed making animated pie chart for", series_name)
```
